```python
import pandas as pd
import urllib
import json
import time

# Initialize the article DataFrame
df1000 = pd.DataFrame(columns=['headline_main', 'pub_date','lead_paragraph'])

# Testing with a single year
for year in range(2020, 2025):  # for testing
    year_begin = str(year) + '0101'
    year_end = str(year) + '1231'
    print("Current year:", year)
    
    for p in range(0, 2):  # for testing, two pages
        print("Current page:", p)
        
        para_dict = {
            "api-key": apikey,
            "q": "politics",
            "begin_date": year_begin,
            "end_date": year_end,
            "page": p
        }
        
        # Build the full URL
        url2check = baseurl + urllib.parse.urlencode(para_dict)
        
        # Send request and retrieve result
        result = urllib.request.urlopen(url2check).read()
        
        # Parse the result as a JSON object
        result_dict = json.loads(result)
        
        # Normalize the JSON response to a DataFrame
        df10 = pd.json_normalize(result_dict['response']['docs'], sep='_')
        
        # Print columns to check the structure
        print(f"Columns in df10 for year {year}, page {p}: {df10.columns}")
        
        # Check if expected columns exist
        expected_columns = ['headline_main', 'pub_date',]
        missing_columns = [col for col in expected_columns if col not in df10.columns]
        if missing_columns:
            print(f"Missing columns: {missing_columns}")
        else:
            # Select the required columns if they exist
            df10 = df10[expected_columns]
        
        # Append the data to the main DataFrame
        df1000 = pd.concat([df1000, df10], ignore_index=True)
        
        # Pause for API usage limits
        time.sleep(12)

# Check the resulting DataFrame
#print(df1000.head())

# output df to a csv file, ignore the index column
df1000.to_csv('cryptoapi4.csv', index=False)
```

```python
import pandas as pd

sms = pd.read_csv('cryptoapi3.csv')

sms.head(10)
```

```python
corpus = sms['headline_main']
corpus.head()
```

```python
# import nltk vader library
from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
# initiate an analyzer
sia = SentimentIntensityAnalyzer()

senti_pos = []
senti_neg = []
senti_neu = []
senti_comp = []


# iterate through each sentence in corpus
for sentence in corpus:
    
    #print(sentence)
    
    # analyze the sentiment. ss is a dictionary
    ss = sia.polarity_scores(sentence)
    
    # output each sentiment score (neg, neu, pos, compound) in ss
    #print(ss['pos']) # for debugging
    senti_pos.append(ss['pos'])
    senti_neg.append(ss['neg'])
    senti_neu.append(ss['neu'])
    senti_comp.append(ss['compound'])
    
    # print an empty line as seperator
    #print('\n')
```

```python
# adding the list to the dataframe as column using assign(column_name = data)
sms = sms.assign(pos = senti_pos, neg = senti_neg)
sms = sms.assign(neg = senti_neg)
sms = sms.assign(neu = senti_neu)
sms = sms.assign(compound = senti_comp)
```

```python
X = sms[['pos','neg', 'neu', 'compound']]
```
